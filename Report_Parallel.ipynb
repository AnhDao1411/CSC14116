### Thiết kế


Kích thước của ma trận đặc trưng: $(1,w,h,d)$ \
**Activation("relu")** \
Công thức: $f(x) = max(0,x)$

*  Sử dụng 3 vòng lặp tương ứng với depth, width, height để duyệt ma trận và gán các giá trị < 0 thành 0.

**MaxPool2D** \
MaxPool2D là layer dùng để giảm kích thước của đặc trưng. Nhóm tham khảo hướng giải quyết thông qua bài viết [1]

*   Tính kích thước và khởi tạo ma trận output:
 * $inputWidth = \frac{inputWidth - poolsize}{stride}+1$
 * $outputWidth = \frac{outputWidth - poolsize}{stride}+1$
*   Duyệt từng d của feature map
  - Sử dụng 2 dòng for để duyệt ma trận và áp dụng maxpool filter. Kết quả được cập nhật trong ma trận output.

**Convolution 2D**

Convolution (lớp tích chập) là lớp trích xuất các đặc tính từ hình ảnh. Tham số lớp này bao gồm một tập hợp các bộ lọc đã được học. Các kernel có kích thước hai chiều đầu tiên đa số 3x3 và có độ sâu bằng với độ sâu của đầu vào. Bằng cách trượt dần bộ lọc theo chiều ngang và dọc trên ảnh, chúng ta sẽ thu được một feature map chứa các đặc trưng được trích xuất từ trên hình ảnh đầu vào.

**Transpose Convolution 2D**

Lớp Tranpose Convolution là quá trình ngược lại của lớp Convolution. \
* Convolution: biến đổi từ ảnh sang đặc trưng
* Tranpose Convolution: biến đổi từ đặc trưng sang các pixel ảnh

Ngoài ra, có thể gia tăng kích thước đặc trưng một cách đơn giản thông qua Upsampling nhưng có thể gây mất mát thông tin. 


Các bước thực hiện:
1.   Tính các tham số z, p', s'
2.   Chèn z các số 0 giữa các hàng và cột
3.   Thêm p' các số 0 để làm padding
4.   Thực hiện phép tích chập với s'=1

**Batch Normalization**

Lớp Batch Normalization dùng để chuẩn hóa các giá trị của đặc trưng. Nhóm thao khảo công thức của [layer keras](https://keras.io/api/layers/normalization_layers/batch_normalization/)

Công thức: $ \frac{\gamma * (batch - moving\_mean)}{\sqrt{moving\_var + \epsilon}} + \beta $ \

* $\gamma$ : hệ số scaling được cập nhật trong quá trình học 
* $\beta$ : hệ số offset (xê dịch)




**Concatenation**

Lớp Concatenation dùng để tạo skip-connection giúp cho việc truyền gradient về các lớp đầu tốt hơn. Cải thiện quá trình học của mô hình. 
\
Bước thực hiện:
1. Cắt skip-features (đặc trưng ở quá trình downsampling) 
2. Ghép skip-features với các đặc trưng hiện tại

```python
@jit(cache=True)
def relu_v1(x):
  for d in range(x.shape[3]):
    for i in range(x.shape[1]):
      for j in range(x.shape[1]):
        if x[:,i,j,d]<0:
          x[:,i,j,d]=0
  return x
```

#### Activation Sigmoid

```python
@jit(cache=True)
def sig_v1(x):
  for d in range(x.shape[3]):
    for i in range(x.shape[1]):
      for j in range(x.shape[1]):
        x[:,i,j,d]= 1 / (1 + np.exp(-x[:,i,j,d]))
  return x 
```

#### MaxPool2D

```python
@jit(cache=True)
def maxpool2d_v1(x, pool_size = 2,stride = 2):
  # Input dim (1,256,256,3)
  output_width = output_height = int((x.shape[1]-pool_size)/stride)+1
  pooled = np.zeros((1,output_width,output_height,x.shape[3]))

  # Iterate over all pools
  for d in range(x.shape[3]):
    for i in range(0,x.shape[1], stride):
      for j in range(0, x.shape[1], stride):
        if i+pool_size <= x.shape[1] and j+pool_size <= x.shape[1]:
          mat = x[:,i:i+pool_size, j:j+pool_size, d]
          pooled[:,int(i/stride),int(j/stride),d]=np.max(mat)
        else:
          raise Exception("Out of index") 
      
  return pooled
```

#### Conv2D

```python
@jit(cache=True)
def conv2D_v1(input, wei, bias, padding='none', stride=1):
    w_in, h_in, c_in = input.shape[1], input.shape[2], input.shape[3]
    k = wei.shape[1] # kernel size
    pad = int(k / 2) if (padding == 'same') else 0
    w_out = (w_in + 2 * pad - k)//stride + 1
    h_out = (h_in + 2 * pad - k)//stride + 1
    c_out = bias.shape[0]

    input_pad = np.zeros((1, w_in + 2 * pad, h_in + 2 * pad, c_in))
    input_pad[:, pad:(w_in + pad), pad:(h_in + pad), :] = input[:, :, :, :]

    res=np.zeros((1, w_out, h_out, c_out))

    for x in range(w_out): 
        for y in range(h_out):
            for c_o in range(c_out):
                total=0
                for c_i in range(c_in):
                    kt=0
                    for k1 in range(k):
                        for k2 in range(k):
                            pos1 = k1 + x * stride
                            pos2 = k2 + y * stride
                            kt += wei[c_o, k1, k2, c_i] * input_pad[0, pos1, pos2, c_i]
                    total += kt
                res[0, x, y, c_o] = total + bias[c_o]
    return res
```

#### Conv2D Transpose

```python
@jit(cache=True)
def trans_conv2D_v1(input, wei, bias, padding=0, stride=1):
    w_in, h_in, c_in = input.shape[1], input.shape[2], input.shape[3]
    k = wei.shape[1] # kernel size
    if padding == "same":
      pad_trans = k-1
    else: pad_trans = k-padding-1
      
    c_out = bias.shape[0]

    # Add zero between rows and columns
    z = stride-1
    input_stride = np.zeros((1, w_in + z*(w_in-1), h_in + z*(h_in-1), c_in))
    for x in range(w_in): 
      for y in range(h_in):
        input_stride[:,x+x*z, y+y*z,:] = input[:,x,y,:]
    input_pad = np.zeros((1, input_stride.shape[1] + 2 * pad_trans, input_stride.shape[2] + 2 * pad_trans, c_in))
    input_pad[:, pad_trans:(input_stride.shape[1] + pad_trans), pad_trans:(input_stride.shape[2] + pad_trans), :] = input_stride[:, :, :, :]
    
    # Calculate output_width/height - no output_padding
    if padding == "same":
      w_out = w_in*stride
      h_out = h_in*stride
    else:
      w_out = (w_in-1)*stride + k - 2*padding
      h_out = (h_in-1)*stride + k - 2*padding

    res=np.zeros((1, w_out, h_out, c_out))
    for x in range(w_out): 
        for y in range(h_out):
            for c_o in range(c_out):
                total=0
                for c_i in range(c_in):
                    kt=0
                    for k1 in range(k):
                        for k2 in range(k):
                            pos1 = k1 + x 
                            pos2 = k2 + y 
                            kt += wei[c_o, k1, k2, c_i] * input_pad[0, pos1, pos2, c_i]
                    total += kt
                res[0, x, y, c_o] = total + bias[c_o]
    return res
```

#### Batch Normalization

```python
@jit(cache=True)
def batch_norm_v1(X, gamma, beta, moving_mean, moving_var):
  eps = 0.001
  Y = X.copy()
  for d in range(X.shape[3]):
    for i in range(X.shape[1]):
      for j in range(X.shape[1]):
        Y[:,i,j,d] = (gamma[d]*(Y[:,i,j,d] - moving_mean[d]) / np.sqrt(moving_var[d] + eps))+ beta[d]
  return Y
```

#### Concatenation

```python
@jit(cache=True)
def concat_v1(skip_features, x):
  w_in, h_in, c_in = x.shape[1] ,x.shape[2] ,x.shape[3]
  Y = np.zeros((1,w_in, h_in, c_in*2))
  crop_pix = (skip_features.shape[1]-w_in)//2
  Y[:, :, :, c_in:] = skip_features[:, crop_pix:(w_in + crop_pix), crop_pix:(h_in + crop_pix), :]
  Y[:, :, :, :c_in] = x[:, :, :, :]
  return Y
```

#### Mô hình U-Net

```python
# load weight to dictionary
weights = {}
for i, layer in enumerate(model.layers):
    if "conv2d_transpose" in layer.name: 
        weights[i] = {"name":layer.name, "wei": layer.weights[0].numpy().transpose((2, 0, 1, 3)),"bias": layer.weights[1].numpy()} 
    elif "conv2d" in layer.name: 
      weights[i] = {"name":layer.name, "wei": layer.weights[0].numpy().transpose((3, 0, 1, 2)),"bias": layer.weights[1].numpy()}
    elif "batch_normalization" in layer.name:
      weights[i] = {"name":layer.name, "gamma":layer.weights[0].numpy(),"beta":layer.weights[1].numpy(),"moving_mean":layer.weights[2].numpy(),"moving_var":layer.weights[3].numpy()  }
      
```

```python
for key,val  in weights.items():
  if "wei" in val.keys():
    print(key,val["name"],val["wei"].shape) 
  else:
    print(key,val["name"])
```

```python
def conv_block(inputs, index):
    x = conv2D_v1(inputs, weights[index]["wei"], weights[index]["bias"], padding='same')
    x = batch_norm_v1(x, weights[index+1]["gamma"], weights[index+1]["beta"], weights[index+1]["moving_mean"], weights[index+1]["moving_var"])
    x = relu_v1(x)

    index += 3
    x = conv2D_v1(x, weights[index]["wei"], weights[index]["bias"], padding='same')
    x = batch_norm_v1(x, weights[index+1]["gamma"], weights[index+1]["beta"], weights[index+1]["moving_mean"], weights[index+1]["moving_var"])
    x = relu_v1(x)
    return x
```

```python
def encoder_block(inputs, index):
    print("encoder_block",index)
    x = conv_block(inputs, index)
    p = maxpool2d_v1(x, 2, 2)
    return x, p
```

```python
def decoder_block(inputs, skip_features, index):
    print("decoder_block",index)
    x = trans_conv2D_v1(inputs, weights[index]["wei"], weights[index]["bias"], padding="same",stride=2)
    x = concat_v1(skip_features, x)
    x = conv_block(x, index + 2)
    return x
```

```python
def build_unet(X):
    index = 1
    """ Encoder """
    s1, p1 = encoder_block(X, index) #64
    index +=7
    s2, p2 = encoder_block(p1,index ) #128
    index +=7
    s3, p3 = encoder_block(p2,index ) #256
    index +=7
    s4, p4 = encoder_block(p3,index ) #512
    index +=7
    b1 = conv_block(p4, index) #1024
    index +=6
    """ Decoder """
    d1 = decoder_block(b1, s4, index) #512
    index +=8
    d2 = decoder_block(d1, s3, index) #256
    index +=8
    d3 = decoder_block(d2, s2, index) #128
    index +=8
    d4 = decoder_block(d3, s1, index) #64 
    index +=8

    """ Output """
    # outputs = Conv2D(1, (1, 1), padding="same", activation="sigmoid")(d4)
    flatten = conv2D_v1(d4, weights[index]["wei"], weights[index]["bias"], padding="same",stride = 1)
    outputs = sig_v1(flatten)
    return outputs
```

```python
@cuda.jit(debug=True) 
def relu_parallel(input, output):
    x, y, z = cuda.grid(3)

    if x < output.shape[1] and y < output.shape[2] and z < output.shape[3]:
        if input[0, x, y, z] < 0: 
            output[0, x, y, z] = 0
        else: output[0, x, y, z] = input[0, x, y, z]
```

#### MaxPool Parallel

```python
@cuda.jit(debug=True)
def maxpool2d_parallel(input, output, pool, stride):
    x, y, z = cuda.grid(3)

    if x < output.shape[1] and y < output.shape[2] and z < output.shape[3]:
        col = x * stride
        row = y * stride

        max = input[0, col, row, z]
        for i in range(pool):
            for j in range(pool):
                if input[0, col + i, row + j, z] > max:
                    max = input[0, col + i, row + j, z]

        output[0, x, y, z] = max
```

```python
before_maxpool = np.array(layer_outs[6])
after_maxpool = np.array(layer_outs[7])
pool = 2
stride = 2

w_out = (before_maxpool.shape[1] - pool)//stride + 1
h_out = (before_maxpool.shape[2] - pool)//stride + 1
pooled = np.zeros((1, w_out, h_out, before_maxpool.shape[3]))

%time maxpool2d_parallel[GRID_SIZE, BLOCK_SIZE](before_maxpool, pooled, pool, stride)
print("Checking difference",np.abs(pooled-after_maxpool).sum())
```
